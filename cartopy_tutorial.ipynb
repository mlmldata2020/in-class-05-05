### Adding latitude and longitude ticks

```python
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER

plt.figure()
ax = plt.axes(projection=ccrs.Mercator())
gl = ax.gridlines(draw_labels=True)
gl.xformatter = LONGITUDE_FORMATTER
gl.yformatter = LATITUDE_FORMATTER
ax.coastlines()
```

The `make_map` function in the `maptools.py` module can be used to execute this code auotmatically.

```python
from maptools import make_map # requires maptools.py!

plt.figure()
ax = make_map(ccrs.Mercator())
ax.coastlines()
```
